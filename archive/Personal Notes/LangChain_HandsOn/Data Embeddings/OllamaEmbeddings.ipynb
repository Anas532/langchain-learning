# Ollama Embeddings — Local Setup

## Overview

This setup uses **Ollama running locally** to generate embeddings without any external API calls.  
The model used is `embeddinggemma:latest`, which is served from:

http://ollama.com/search?c=embedding

No API key is required because everything runs locally.

---

## What is an Embedding?

An embedding is a numerical vector representation of text.

It converts natural language into numbers so that:
- semantic similarity can be measured
- vector databases can store text meaningfully
- retrieval-augmented generation (RAG) can work
- search can become semantic instead of keyword-based

In this setup, each embedding has **768 dimensions**.

---

## embed_documents()

Used when embedding multiple texts at once.

Input:
- A list of strings

Output:
- A list of vectors
- One vector per input string

Purpose:
- Used when storing documents into a vector database
- Used during indexing phase of RAG

Example use case:
Embedding paragraphs before storing them in FAISS or Chroma.

---

## embed_query()

Used when embedding a single string.

Input:
- A single string

Output:
- A single vector

Purpose:
- Used during search
- Converts user query into a vector
- Compared against stored document vectors

Example use case:
User asks: “What is the second letter of Greek alphabet?”
The query is embedded and matched against stored document vectors.

---

## Conceptual Difference

embed_documents() → For indexing data  
embed_query() → For searching against indexed data  

Both return vectors of the same dimension.

---

## Why This Is Useful

Using local embeddings with Ollama:
- Avoids API costs
- Avoids rate limits
- Keeps data private
- Works offline

This setup is ideal for learning RAG pipelines and vector similarity systems.


```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = OllamaEmbeddings(model='embeddinggemma:latest')

```

```python
embeddings
```

```python
r1 = embeddings.embed_documents([
    "This is a test",
    "This is another test",
    "This is yet another test"
])
```

```python
print(len(r1[0]))
```

```python
query = "What is the second letter of Greek alphabet?"

query_vector = embeddings.embed_query(query)
```
