# HuggingFace Embeddings — Local Model

## Overview

This setup uses a **local HuggingFace sentence-transformer model** to generate embeddings without relying on external APIs.

Model used:
`all-MiniLM-L6-v2`

This model runs locally through the `sentence-transformers` backend.

No OpenAI key is required.

---

## What This Model Does

The model converts text into a dense numerical vector representation.

Each text input is transformed into a fixed-size embedding vector that captures semantic meaning.

These embeddings are typically:
- 384 dimensions (for all-MiniLM-L6-v2)
- Lightweight
- Fast
- Suitable for semantic search and RAG

---

## Environment Setup

Environment variables may be loaded using `.env`, but for this specific model:

- A HuggingFace token is optional
- Required only for private models or API usage
- Not required for local sentence-transformers

---

## Embedding Flow

### 1. Model Initialization

The embedding model is initialized once.

This loads the transformer model into memory.

---

### 2. embed_query()

Used when embedding a single string.

Input:
- One string

Output:
- One vector

Used for:
- Query embeddings
- Semantic search comparison
- Retrieval phase in RAG

---

### 3. embed_documents()

Used when embedding multiple texts.

Input:
- List of strings

Output:
- List of vectors

Used for:
- Indexing documents
- Storing vectors in a database
- Building vector stores

---

## Key Characteristics

- Runs locally
- No rate limits
- No API cost
- Lower dimensionality than some larger embedding models
- Good balance of speed and performance

---

## When to Use This Setup

Use HuggingFace local embeddings when:
- You want zero API cost
- You are experimenting or learning RAG
- You need privacy
- You want fast prototyping

```python
text = 'This is a test document to be embedded.'
query_result = embeddings.embed_query(text)
```
